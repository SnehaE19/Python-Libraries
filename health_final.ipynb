# ***Health and sleep***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
health=pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
```

```python
health
```

```python
health.info()
```

# **Feature Engineering**

```python
health['Age'].unique()
```

```python
health['age_group']=pd.cut(health['Age'],bins=[25,40,60],labels=['adults','senior'])
```

```python
health
```

# **Distribution of Sleep Duration**

```python
plt.figure(figsize=(8, 5))
sns.histplot(health["Sleep Duration"], bins=20, kde=True, color="blue")
plt.title("Distribution of Sleep Duration")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Count")
plt.show()
```

*The distribution is slightly right-skewed meaning some individuals sleep for fewer hours but most fall within a normal range of 5-8 hours.*

# **Physical Activity and Sleep Duration**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Physical Activity Level", y="Sleep Duration", data=health, hue="Gender", alpha=0.7)
plt.title("Physical Activity vs. Sleep Duration")
plt.xlabel("Physical Activity Level")
plt.ylabel("Sleep Duration (hours)")
plt.show()
```

*There is no strong correlation, but individuals with higher physical activity levels tend to have slightly longer sleep durations*

# **Sleep Duration and Age**

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="Age",y="Sleep Duration",data=health,scatter_kws={"alpha":0.5},line_kws={"color":"green"})
plt.title("Age vs. Sleep Duration")
plt.xlabel("Age")
plt.ylabel("Sleep Duration (hours)")
plt.show()
```

*There is a slight incline in sleep duration with increasing age.*

# **Gender and Sleep Duration**

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
obj=sns.barplot(data=health,x="Gender",y='Sleep Duration',palette='coolwarm')
for i in obj.containers:
  obj.bar_label(i)
plt.grid()

plt.subplot(1,2,2)
sns.boxplot(data=health,x="Gender",y='Sleep Duration',palette='coolwarm')

plt.grid()
```

*While comparing Gender and Sleep Duration we can see that Female is having more sleep duration*

# **Gender and quality of sleep**

```python
obj=sns.violinplot(x="Gender",y='Quality of Sleep',data=health,palette="viridis")
# for i in obj.containers:
#   obj.bar_label(i)
obj
plt.grid()
```

*Comparitively Quality of sleep is also more for females*

# **Gender and Physical Activity Level**

```python
sns.pointplot(x="Gender", y="Physical Activity Level", data=health,palette='magma')
plt.grid()
plt.show()
```

*Physical activity is more practised by Male*


# **Stress Level based on sleep Duration and Sleep Quality**

```python
plt.figure(figsize=(12,5))
plt.subplot(2,1,1)
sns.lmplot(x="Sleep Duration", y="Stress Level", data=health,palette="rainbow")
plt.title("Sleep Duration vs. Stress Level")
plt.xlabel("Sleep Duration")
plt.ylabel("Stress Level")
plt.show()

plt.subplot(2,1,2)
plt.figure(figsize=(8, 5))
sns.boxplot(x="Quality of Sleep", y="Stress Level",data=health,palette="Spectral")
plt.title("Sleep Quality vs. Stress Level")
plt.xlabel("Quality of Sleep")
plt.ylabel("Stress Level")
plt.show()
```

*Higher stress levels are associated with lower sleep quality.People with low sleep quality tend to have high stress scores.*

# **Daily Steps Distribution**

```python
plt.figure(figsize=(8, 5))
plt.subplot(1,2,1)
sns.histplot(data=health,x="Daily Steps",bins=20,kde=True,color='orange')
plt.title("Distribution of Daily Steps")
plt.xlabel("Daily Steps")
plt.ylabel("Count")
plt.show()
plt.subplot(1,2,2)
sns.regplot(x="Daily Steps",y="Stress Level", data=health, scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
plt.title("Daily Steps vs. Stress Level")
plt.xlabel("Daily Steps")
plt.ylabel("Stress Level")
plt.show()

```

*Most individuals take between 3,000 to 10,000 steps daily.There is a slight negative correlation, meaning people who walk more tend to have lower stress levels.*

# **Distribution of BMI Categories**

```python
plt.figure(figsize=(7, 7))
bmi_counts = health["BMI Category"].value_counts()
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', colors=["lightgreen", "yellow", "salmon","red"], startangle=140)
plt.title("Distribution of BMI Categories")
plt.show()
```

# **Correlation**

```python
health1=pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
health1.set_index('Person ID',inplace=True)
```

```python
health1['Gender']=health1['Gender'].map({'Male':0,'Female':1,'Other':2})
health1['Occupation']=health1['Occupation'].map({'Software Engineer':0,'Doctor':1,'Sales Representative	':2,'Nurse':3,'Teacher':4,'Lawyer':5, 'Engineer':6, 'Accountant':7, 'Scientist':8,
       'Salesperson':9, 'Manager':10})
health1['BMI Category']=health1['BMI Category'].map({'Overweight':0,'Normal':1,'Obese':2,'Normal Weight':3})
health1['Sleep Disorder']=health1['Sleep Disorder'].map({'NaN':0,'Sleep Apnea':1,'Insomnia':2})
```

```python
health1
```

# **Heatmap**

```python
data=health1.corr(numeric_only=True)
data
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=data,annot=True,linewidth=1,linecolor='pink',cmap='icefire',fmt='.3f')
```
